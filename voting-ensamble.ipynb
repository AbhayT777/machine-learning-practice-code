```python
import pandas as pd
```

```python
df = pd.read_csv('data/Iris.csv')
df
```

```python
df = df.iloc[:,1:]
df
```

```python
from sklearn.preprocessing import LabelEncoder
LB = LabelEncoder()
df['Species'] = LB.fit_transform(df['Species'])
df
```

```python
# import seaborn as sns
# sns.pairplot(df, hue= 'Species')
```

```python
df = df[df['Species'] != 0][['SepalLengthCm', 'SepalWidthCm', 'Species']]
df
```

```python
x = df.iloc[:,0:2]
y = df.iloc[:,-1]
y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
c1 = LogisticRegression()
c2 = KNeighborsClassifier()
c3 = RandomForestClassifier()
```

```python
estimators = [('lr', c1), ('kn', c2), ('rf', c3)] #list of tuples having algorithm ka naam and object of that algo
```

```python
for name, model in estimators:
    scores = cross_val_score(model,x,y,cv=10, scoring= 'accuracy')
    print(name , np.round(np.mean(scores), 2))
```

## Approach 1:

Hard Voting

```python
from sklearn.ensemble import VotingClassifier
vc = VotingClassifier(estimators= estimators, voting= 'hard')
score = cross_val_score(vc, x, y, cv= 10, scoring= 'accuracy')
print(np.round(np.mean(score), 2))
```

Soft Voting

```python
vc = VotingClassifier(estimators= estimators, voting= 'soft')
score = cross_val_score(vc, x, y, cv= 10, scoring= 'accuracy')
print(np.round(np.mean(score), 2))
```

Weighted Voting

```python
for i in range(1,4):
    for j in range(1,4):
        for k in range(1,4):
            vc = VotingClassifier(estimators= estimators, voting= 'soft', weights= [i,j,k])
            score = cross_val_score(vc, x, y, cv= 10, scoring= 'accuracy')
            print('for i={}, j={}, k={}'.format(i,j,k), np.round(np.mean(score), 2))
```

## Approach 2: Classifiers of same Algo

```python
from sklearn.svm import SVC
from sklearn.datasets import make_classification
X, Y = make_classification(n_samples= 1000, n_features= 20, n_informative= 15, n_redundant= 5, random_state= 2)

svm1 = SVC(probability= True, kernel= 'poly', degree= 1)
svm2 = SVC(probability= True, kernel= 'poly', degree= 2)
svm3 = SVC(probability= True, kernel= 'poly', degree= 3)
svm4 = SVC(probability= True, kernel= 'poly', degree= 4)

estimators = [('svm1', svm1), ('svm2', svm2), ('svm3', svm3), ('svm4', svm4)]

for est in estimators:
    score = cross_val_score(est[1], x, y, cv= 10, scoring= 'accuracy')
    print(est[0], np.round(np.mean(score), 2))
```

```python
vc1 = VotingClassifier(estimators= estimators, voting= 'soft')
score = cross_val_score(vc1, x, y, cv= 10, scoring= 'accuracy')
print(np.round(np.mean(score), 2))
```
