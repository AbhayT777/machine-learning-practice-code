```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[1,2], [3,4], [2,4], [7,8], [9,9], [25,80]])
```

```python
plt.scatter(x[:,0], x[:,1])
plt.show
```

```python
db = DBSCAN(eps= 10.0, min_samples= 2)
db.fit(x)
db.labels_
```

```python
from sklearn.datasets import make_circles
d, _ = make_circles(n_samples= 500, factor= .5, noise= 0.03, random_state= 4)

dbs= DBSCAN(eps= 0.1, min_samples= 5)
clus = dbs.fit_predict(d)

plt.scatter(d[:,0], d[:,1], c= clus, cmap= 'viridis')
plt.show
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters= 3)

clus2 = km.fit_predict(d)

plt.scatter(d[:,0], d[:,1], c= clus2, cmap= 'viridis')
plt.show
```
